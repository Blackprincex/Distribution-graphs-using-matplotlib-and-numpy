```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# scatter diagram

x1 = 1.5*np.random.randn(150) + 10
y1 = 1.5*np.random.randn(150) + 10

x2 = 1.5*np.random.randn(100) + 4
y2 = 1.5*np.random.randn(100) + 4

x = np.append(x1,x2)
y = np.append(y1,y2)

fig, ax = plt.subplots()
ax.scatter(x,y)
plt.show()
```
